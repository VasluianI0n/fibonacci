```python
import matplotlib.pyplot as plt
import time


fibon_num = []
x = []
y = []
iterations = 0
def fib(n):
    global iterations
    new, old = 1, 0
    for i in range(n):
            new, old = old, new + old
            iterations+=1 #keep track of how many times its run
    return old


for i in range(40):
    start = time.time()
    fibon_num.append(fib(i))
    end = time.time()
    x.append(i)
    y.append(end-start)
    print(i,'Fibonacci term= ',fibon_num[i],' and was computed in ', end-start, ' seconds')
    plt.plot(x,y)
    plt.draw()
    
plt.xlabel('x - n-th Fibonacci number')
plt.ylabel('y - time spent (s)')
plt.show()
print("sum:", fib(40)) #sum:102334155
print("total iterations:", iterations) #total iterations 40
```

```python
import numpy as np
from numpy.linalg import matrix_power
import matplotlib.pyplot as plt
import time


fibon_num = []
x = []
y = []

iterations = 0
def fibmat(n):
    global iterations
    i = np.array([[0, 1], [1, 1]])
    iterations+=1
    return np.matmul(matrix_power(i, n), np.array([1, 0]))[1]

for i in range(40):
    start = time.time()
    fibon_num.append(fibmat(i))
    end = time.time()
    x.append(i)
    y.append(end-start)
    print(i,'Fibonacci term= ',fibon_num[i],' and was computed in ', end-start, ' seconds')
    plt.plot(x,y)
    plt.draw()

plt.xlabel('x - n-th Fibonacci number')
plt.ylabel('y - time spent (s)')
plt.show()



print(fibmat(40)) #output : 102334155
```

```python
import matplotlib.pyplot as plt
import time

fibon_num = []
x = []
y = []

def fib(n):
    if n%2 == 0:#even
        return fib_even(n)
    return fib_odd(n)

def fib_odd(N):
    n = int((N+1)/2)
    Fn = fibmat(n) 
    Fn1 = fibmat(n-1)
    return Fn1**2 + Fn**2

def fib_even(N):
    n = int(N/2)
    Fn = fibmat(n) 
    Fn1 = fibmat(n-1)
    return Fn * (2*Fn1 + Fn)

for i in range(40):
    start = time.time()
    fibon_num.append(fib(i))
    end = time.time()
    x.append(i)
    y.append(end-start)
    print(i,'Fibonacci term= ',fibon_num[i],' and was computed in ', end-start, ' seconds')
    plt.plot(x,y)
    plt.draw()
    

plt.xlabel('x - n-th Fibonacci number')
plt.ylabel('y - time spent (s)')
plt.show()

print(fib(40)) #op: 102334155
print(fib(39)) #op: 63245986
```

```python
import matplotlib.pyplot as plt
import time

fibon_num = []
x = []
y = []


def nth_fibonnaci(n):
    if n > 0:
        n1, n2 = 1, 1
        if n > 3:
            for _ in range((n//3)):
                n1, n2 = n2, (n2 << 2)+n1  # << 2   is multiply by 4
        if n % 3 == 0:
            return n1
        elif n % 3 == 1:
            return (n2-n1) >> 1  # >> 1   is divide by 2  'F1'
        elif n % 3 == 2:
            return (n2+n1) >> 1  # >> 1   is divide by 2  'F2'
    else:
        return -1
 
 
 
for i in range(40):
    start = time.time()
    fibon_num.append(nth_fibonnaci(i+1))
    end = time.time()
    x.append(i)
    y.append(end-start)
    print(i,'Fibonacci term= ',fibon_num[i],' and was computed in ', end-start, ' seconds')
    plt.plot(x,y)
    plt.draw()
    

plt.xlabel('x - n-th Fibonacci number')
plt.ylabel('y - time spent (s)')
plt.show()
```
